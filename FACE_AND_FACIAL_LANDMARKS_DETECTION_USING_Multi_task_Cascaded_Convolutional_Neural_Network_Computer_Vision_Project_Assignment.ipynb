```python
from mtcnn import MTCNN
```

```python
model=MTCNN()
```

```python
import cv2
```

```python
image=cv2.imread('/content/image.jpg')
```

```python
List_of_dic=model.detect_faces(image)
```

```python
List_of_dic
```

```python
List_of_dic[0]['keypoints']
List_of_dic[0]
```

```python
for face in List_of_dic:
  cv2.rectangle(image,pt1=(face['box'][0],face['box'][1]),pt2=(face['box'][0]+face['box'][2],face['box'][1]+face['box'][3]),color=(255,255,0),thickness=3)
  cv2.putText(image,text=f'confidence: {face['confidence']:0.3f}',color=(255,255,0),org=(face['box'][0],face['box'][1]-3),fontFace=0,fontScale=1)
  cv2.circle(image,center=(face['keypoints']['nose'][0],face['keypoints']['nose'][1]),color=(255,0,0),radius=5)
  cv2.circle(image,center=(face['keypoints']['right_eye'][0],face['keypoints']['right_eye'][1]),color=(0,255,255),radius=5)
  cv2.circle(image,center=(face['keypoints']['left_eye'][0],face['keypoints']['left_eye'][1]),color=(0,255,255),radius=5)
  cv2.circle(image,center=(face['keypoints']['mouth_right'][0],face['keypoints']['mouth_right'][1]),color=(0,255,255),radius=5)
  cv2.circle(image,center=(face['keypoints']['mouth_left'][0],face['keypoints']['mouth_left'][1]),color=(0,255,255),radius=5)


```
